#### In order to figure out how's temperature varying in different regions of usa create a new column which contains all 5 different regions of usa


```python
# define afunction to determine region
def get_us_region(lat, long):
  if (lat > 38 and lat  <45) and (long >-75 and long <-70):
    return "Northeast"
  elif (lat > 30 and lat < 38) and (long > -85 and long < -80):
    return "Southeast"
  elif (lat > 36.15 and lat < 49.5) and (long > -105.5 and long < -79):
    return "Midwest"
  elif (lat > 29 and lat < 39 ) and (long > -117 and long < -104):
    return "Southwest"
  elif (lat > 7 and lat < 85 ) and (long > - 179 and long  < -20) :
    return "West"
  else:
    return "Unknown"
```

```python
accident_df['Regions']  = accident_df.apply(lambda x :get_us_region(x.Start_Lat, x.Start_Lng), axis= 1)
```

```python
#import neccesary  Libraries
import astral
import datetime
```

```python
#Extract date from start_time
accident_df['Date'] = accident_df['Start_Time'].dt.date
```

```python
#define a function to get sunrise and sunset time
def get_sunrise_sunset(city,state,tz,lat,lng,dat):
  from astral import LocationInfo
  loc = LocationInfo(name=city, region=state, timezone=tz,
                   latitude=lat, longitude=lng	)
  from astral.sun import sun
  s = sun(loc.observer, date=dat, tzinfo=loc.timezone)
  return s['sunrise'].strftime('%H:%M %p'), s['sunset'].strftime('%H:%M %p')
```

```python
# this  cell will take 2 minutes to run
accident_df['Sunrise_Time'], accident_df['Sunset_Time'] = zip(*accident_df.apply(lambda x: get_sunrise_sunset(x.City, x.State, x.Timezone, x.Start_Lat, x.Start_Lng,x.Date), axis=1))
```
